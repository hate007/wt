```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.probability import FreqDist
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text_paragraph = """Mai HU gyiaaan maera gana hai surila. YOu are noob, I'm pro, but who know what will be the future.
Lets embrace it, and then face it because we dont know what's there for us. SO lets runnnnnnnn, runnn runnn from
the procarsination. LOl dont run from your problems"""

nltk.download('punkt')
nltk.download('stopwords')

tokens = word_tokenize(text_paragraph)

stop_words = set(stopwords.words('english'))
filtered_text = [word for word in tokens if word.lower() not in stop_words]

word_freq_dist = FreqDist(filtered_text)

plt.figure(figsize=(10, 5))
word_freq_dist.plot(20, title='Top 20 Word Frequencies')


wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_text))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud')
plt.show()
```
