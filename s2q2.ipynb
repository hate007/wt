```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = {
    'Salary': [4000, 5000, 6000, 8000, 11000, 15000, 20000, 30000, 50000, 100000],
    'Purchases': [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000],
}

df = pd.DataFrame(data)

X = df[['Salary']] #indpenet - does not depend, change to obeserve
y = df['Purchases']         #depended - predict

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)

print("Training set:")
print(X_train.values)
print(y_train.values)
print("\nTesting set:")
print(X_test.values)
print(y_test.values)

model = LinearRegression()
model.fit(X_train.values, y_train)

# Predicting_to_check_linear_regerssion_model
predicted_purchase = model.predict([[4000]])
print("\nPredicted purchase for salary 900:", predicted_purchase)
```

```python
# Plootting
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X.values), color='red')
plt.title('Salary vs Purchase')
plt.xlabel('Salary')
plt.ylabel('purchase')
plt.yticks(np.linspace(1000, 10000, 10)) #optional
plt.xticks(np.linspace(4000, 100000, 10)) #optional
plt.show()
```
