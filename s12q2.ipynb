```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'Height': [150, 155, 160, 165, 170],
    'Weight': [55, 60, 65, 70, 75]
}

df = pd.DataFrame(data)

X = df[['Height']] #indpenet - does not depend, change to obeserve
y = df['Weight']         #depended - predict

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=42)

print("Training set:")
print(X_train.values)
print(y_train.values)
print("\nTesting set:")
print(X_test.values)
print(y_test.values)

model = LinearRegression()
model.fit(X_train.values, y_train)

predicted_weight = model.predict([[155]])
print("\nPredicted Weight for height 155:", predicted_weight)

# Plootting
plt.scatter(X, y, color='blue')
plt.plot(X, model.predict(X.values), color='red')
plt.title('Height vs Weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```
